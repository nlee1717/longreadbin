```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
flye = pd.read_csv("USA_flye", header=None)[0]
polished = pd.read_csv("USA_nextPolish", header=None)[0]
mega = pd.read_csv("USA_megaHit", header=None)[0]
```

```python
flye = flye.values
flye.sort()
flye = flye[::-1]
polished = polished.values
polished.sort()
polished = polished[::-1]
mega = mega.values
mega.sort()
mega = mega[::-1]
```

```python
fig, ax = plt.subplots(1,1,figsize = (16,10))
ax.plot(flye, np.cumsum(flye),   color = 'green', label = 'Flye', linewidth = 2)
ax.plot(polished, np.cumsum(polished),   color = 'orange', label = 'medaka + nextPolish', linewidth = 2)
ax.plot(mega, np.cumsum(mega),   color = 'blue', label = 'megaHit', linewidth = 2)

ax.set_ylabel('Cumulative No. of Bases Contig')
ax.set_xlabel('Contigs Sorted in Decreasing Order of their Lengths')
ax.legend()
ax.grid()
# ax.set_xscale("log")
ax.set_yscale('log')
ax.set_xlim(ax.get_xlim()[::-1])
ax.axvline(500, color="black")

fig.tight_layout()
fig.savefig('test.pdf')
```
